```python
def remove_outliers(a):
    q25=a.quantile(0.25)
    q75=a.quantile(0.75)
    IQR=q75-q25
    lower_limit=q25-(1.5*IQR)
    upper_limit=q75+(1.5*IQR)
    return a[(a>=lower_limit)&(a<=upper_limit)]
```

```python
def remove_outliers_col(a,col):
    q25=a[col].quantile(0.25)
    q75=a[col].quantile(0.75)
    IQR=q75-q25
    lower_limit=q25-(1.5*IQR)
    upper_limit=q75+(1.5*IQR)
    return a[(a[col]>=lower_limit)&(a[col]<=upper_limit)]
```

```python
cat_columns=["room_type","neighbourhood_group"]
num_columns=["latitude","longitude","number_of_reviews","availability_365","minimum_nights","min_price_pay"]
col_transform=ColumnTransformer(transformers=[
    ("cat",OneHotEncoder(handle_unknown="ignore"),cat_columns),
    ("num",StandardScaler(),num_columns)
],remainder="drop")
```

```python
X,Y=df[cat_columns+num_columns],df["price_log"]
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

def give_color(indep,indep_min,indep_mean,indep_max):
    d1=abs(indep-indep_min)
    d2=abs(indep-indep_mean)
    d3=abs(indep-indep_max)
    if d1<d2:
        if d1<d3:
            return "blue"
        else:
            return "red"
    else:
        if d2<d3:
            return "purple"
        else:
