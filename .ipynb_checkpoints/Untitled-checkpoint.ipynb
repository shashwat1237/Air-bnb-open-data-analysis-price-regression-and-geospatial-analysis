```python
def remove_outliers(a):
    q25=a.quantile(0.25)
    q75=a.quantile(0.75)
    IQR=q75-q25
    lower_limit=q25-(1.5*IQR)
    upper_limit=q75+(1.5*IQR)
    return a[(a>=lower_limit)&(a<=upper_limit)]
```

```python
def remove_outliers_col(a,col):
    q25=a[col].quantile(0.25)
    q75=a[col].quantile(0.75)
    IQR=q75-q25
    lower_limit=q25-(1.5*IQR)
    upper_limit=q75+(1.5*IQR)
    return a[(a[col]>=lower_limit)&(a[col]<=upper_limit)]
```

```python
xgb_reg=xgb.XGBRegressor(objective="reg:squarederror",
                        n_estimators=1000,
                        max_depth=10,
                        random_state=42)
```

```python
cat_columns=["room_type","neighbourhood_group"]
num_columns=["latitude","longitude","number_of_reviews","availability_365","minimum_nights","min_price_pay"]
col_transform=ColumnTransformer(transformers=[
    ("cat",OneHotEncoder(handle_unknown="ignore"),cat_columns),
    ("num",StandardScaler(),num_columns)
],remainder="drop")
```

```python
X,Y=df[cat_columns+num_columns],df["price_log"]
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
pipe=Pipeline(steps=[("preprocessor",col_transform),
                    ("model",xgb_reg)])
param_distributions={"model__n_estimators":[100,200,300,400,500,600,700,800,900,1000],
                   "model__max_depth":[2,3,4,5,6,7,8,9,10]}

grid_search=RandomizedSearchCV(estimator=pipe,
                        cv=5,
                        scoring="r2",
                        n_iter=25,
                        param_distributions=param_distributions)


grid_search.fit(X_train,Y_train)
```

```python
Y_pred=grid_search.predict(X_test)
print(np.sqrt(mean_squared_error(Y_test,Y_pred)))
```

```python
print(r2_score(Y_test,Y_pred))
```

```python
pipe=grid_search.best_estimator_
```

```python
print(pd.Series(pipe.named_steps["model"].feature_importances_,
index=pipe.named_steps["preprocessor"].get_feature_names_out()))
```

```python
X["price_log"]=Y
to_analyze=pipe.named_steps["preprocessor"].transform(X)
inertia_list=[]
for i in range(2,11):
    km=KMeans(n_clusters=i,random_state=42)
    km.fit(to_analyze)
    inertia_list.append(km.inertia_)
sns.lineplot(y=inertia_list,x=list(range(2,11)))
```

```python
km=KMeans(n_clusters=6,random_state=42)
km.fit(to_analyze)
```

```python
print(silhouette_score(to_analyze,km.labels_))
```
